```python
# взять только первый столбец с фотками и последний с артикулами 
df = df[['Image_URL','Textbox_value']]
```

```python
# удалить пропуски 
df.dropna(inplace=True)
```

```python
# сбросить индекс 
df.reset_index(inplace=True)
```

```python
# цикл по всем элементам 

for i in range(df.shape[0]):
    decoded_data= df.loc[i][1].replace('data:image/jpeg;base64,', '').encode('ascii')  # удалить начало base64
                                                                                    # берем данные картинки 
    name_file = f'{df.loc[i][2]}'                                                 # именем файла будет артикул
    name = '/Users/romanvasyliev/Downloads/pics/%s.jpg'%name_file               # путь сохранения файла 
                                 
    print (name)                          # проверка имени файла 
    #img_file = open(name, mode = 'w')          
    try: 
        img_file  = base64.b64decode(decoded_data)     # декодирование
        image_result = open(name, 'wb')         # открыть файл для записи, папка должна быть создана
        image_result.write(img_file)            # записать в файл данные 
    except Exception:                           # если ошибка - вывести 
        print(sys.exc_info()[2])
    
```
